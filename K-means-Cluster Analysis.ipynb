```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
%matplotlib inline
```

```python
#A very simple k-means real-life practice 
#Initialize datasets
data=np.random.rand(1000,2)
#Determine number of clusters
km=KMeans(n_clusters=3).fit(data)
#Label the result, get the categories of the training set
rs_labels=km.labels_
#Get the center of each clsuter
rs_center_ids=km.cluster_centers_
rs_center_ids
```

```python
#Describe each point
plt.scatter(data[:,0],data[:,1],c=rs_labels,alpha=0.5)   #calpha
#Describe the center points
plt.scatter(rs_center_ids[:,0],rs_center_ids[:,1],c='red')
plt.show()
```

```python
#Second simple example
X = np.array([[1, 2], [1, 4], [1, 0],[4, 2], [4, 4], [4, 0]])
#Divide the data into clusters, get the center points, and corresponding cluster
kmeans=KMeans(n_clusters=2,random_state=0).fit(X)
kmeans.labels_
```

```python
#Predict the categories
kmeans.predict([[0,0],[4,4]])
```

```python
#Examine the cluster centers
kmeans.cluster_centers_
```

```python
# X is sample featuresY is cluster categories,total of 1000 samples, each sample has 4 features, total of 4 clusters,cluster centered at [-1,-1], [0,0],[1,1], [2,2], cluster variance [0.4, 0.2, 0.2]
X, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1,-1], [0,0], [1,1], [2,2]], cluster_std=[0.4, 0.2, 0.2, 0.2], 
                  random_state =9)
plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.show()
```
